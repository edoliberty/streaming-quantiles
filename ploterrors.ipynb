```python
import numpy as np

def make_data(n, pattern='rand'):
    np.random.seed(seed=0)
    data=np.zeros(n)
    b = 1/n
    data[0:2] = [0,1]
    if pattern == 'up':
        data[2:] = np.linspace(b, 1-b, n-2)
    elif pattern == 'down':
        data[2:] = np.linspace(1-b, b, n-2)
    elif pattern == 'in':
        data[2::2] = np.linspace(b  , 0.5-b, n//2-1)
        data[3::2] = np.linspace(1-b, 0.5+b, n//2-1)
    elif pattern == 'out':
        data[2::2] = np.linspace(0.5-b, b  , n//2-1)
        data[3::2] = np.linspace(0.5+b, 1-b, n//2-1)
    else:
        data[2:] = np.random.rand(n-2)
    return data
```

```python
n = 10000
data = make_data(n,'down')
```

```python
from stairwaytoheaven import StairwayToHeaven as sketch

errss = []
ks = range(10,101,10)

for k in ks:
    errs = []
    s = sketch(k)
    for i,x in enumerate(data):
        s.update(x)
        errs.append(s.get_error()/(i+1))
    errss.append(errs)
```

```python
import matplotlib.pyplot as plt
for e in errss:
    plt.plot(e)
```
