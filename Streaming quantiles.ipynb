# Plotting results for different K sizes

```python
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 15)
import matplotlib.pyplot as plt
```

```python
import streamMaker
import kll
```

```python
kll_k_16 = kll.KLL(16)
kll_k_32 = kll.KLL(32)
kll_k_64 = kll.KLL(64)
```

```python
streamer = streamMaker.StreamMaker()
n = 1000
order ='zoomin'

for item in streamer.make(n, order):
    kll_k_16.update(item)
    kll_k_32.update(item)
    kll_k_64.update(item)
```

```python
def plot(results, label):
    quantiles = [quantile for item, quantile in results]
    items = [item for item, quantile in results]
    plt.plot(items, quantiles, label=label)

plot(kll_k_16.cdf(), 'k 16')
plot(kll_k_32.cdf(), 'k 32')
plot(kll_k_64.cdf(), 'k 64')

plt.xlabel('Value')
plt.ylabel('Aproximated quantile')
plt.legend()
plt.show()
```
